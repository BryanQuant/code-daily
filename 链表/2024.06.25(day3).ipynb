 
 
 203.移除链表元素 

 时间复杂度: O(n)  空间复杂度: O(1) 

思路: 创建虚拟头节点，如果等于val，就让next跳过下一个节点，指向下下个节点，如果不等于val，直接到下一个节点。

难点：还不够太熟悉链表的next和val指代的是什么，并且不理解null的处理方式。

题目链接/文章讲解/视频讲解：：https://programmercarl.com/0203.%E7%A7%BB%E9%99%A4%E9%93%BE%E8%A1%A8%E5%85%83%E7%B4%A0.html


```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeElements(self, head, val):
        """
        :type head: ListNode
        :type val: int
        :rtype: ListNode
        """
        dummy=ListNode(next=head)
        current=dummy
        while current.next:
            if current.next.val==val:
                current.next=current.next.next
            else:
                current=current.next
        return dummy.next
```

 707.设计链表  

建议： 这是一道考察 链表综合操作的题目，不算容易，可以练一练 使用虚拟头结点

注意：

get：得到index所对应的值,因为虚拟头是自己加的，所以第一个值得是dummy.next，同时最后记得范围current.val而不是current

addAtHead：先让加的node的下一个指针到虚拟头的next，ListNode(val,self.dummy.next)，再让现在虚拟头的next指向新的node

addAtTail：直接循环到最后一个

addAtIndex：index >self.size, 边界是可以到null的

deleteAtIndex：index >self.size, 边界是不可以到null的，直接让current.next=current.next.next

记得所有都要对self.size进行操作

题目链接/文章讲解/视频讲解：https://programmercarl.com/0707.%E8%AE%BE%E8%AE%A1%E9%93%BE%E8%A1%A8.html


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class MyLinkedList(object):

    def __init__(self):
        self.dummy=ListNode()
        self.size=0


    def get(self, index):
        """
        :type index: int
        :rtype: int
        """
        # 判断index在range外
        if index <0 or index >=self.size:
            return -1
        # 得到index所对应的值,因为虚拟头是自己加的，所以第一个值得是dummy.next
        current = self.dummy.next
        for i in range(index):
            current=current.next
        return current.val


    def addAtHead(self, val):
        """
        :type val: int
        :rtype: None
        """
        # 先让加的node的下一个指针到虚拟头的next，再让现在虚拟头的next指向新的node
        self.dummy.next=ListNode(val,self.dummy.next)
        self.size+=1



    def addAtTail(self, val):
        """
        :type val: int
        :rtype: None
        """
        current=self.dummy

        # 直接循环到最后一个
        while current.next:
            current = current.next
        current.next=ListNode(val)
        self.size+=1



    def addAtIndex(self, index, val):
        """
        :type index: int
        :type val: int
        :rtype: None
        """
        if index <0 or index >self.size:
            return 
        current=self.dummy
        for i in range (index):
            current=current.next
        current.next=ListNode(val,current.next)
        self.size+=1

    def deleteAtIndex(self, index):
        """
        :type index: int
        :rtype: None
        """
        if index <0 or index >=self.size:
            return 
        
        current=self.dummy
        for i in range(index):
            current=current.next
        current.next=current.next.next
        self.size-=1
        




# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```

 206.反转链表

注意：得有temp来保存current.next的位置

题目链接/文章讲解/视频讲解：https://programmercarl.com/0206.%E7%BF%BB%E8%BD%AC%E9%93%BE%E8%A1%A8.html 


```python
# 双指针写法
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        pre=None
        current=head
        while current:
            temp=current.next
            current.next=pre
            pre=current
            current=temp
        return pre


```

```python
# 递归写法
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        pre=None
        current=head
        return self.reverse(current,pre)
    def reverse(self,current,pre):
        if current==None:
            return pre
        temp=current.next
        current.next=pre
        return self.reverse(temp,current)
        
```
