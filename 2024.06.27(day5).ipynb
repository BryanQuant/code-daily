 242.有效的字母异位词 

建议： 这道题目，大家可以感受到 数组 用来做哈希表 给我们带来的遍历之处。 

思路：record内取相对位置，也就是要减去ord（'a'）,ord显示的字母位置。s：对每一个字母出现的位置都加上1，然后t出现的再减去1，最后如果不为0说明就有不同的字母出现。

题目链接/文章讲解/视频讲解： https://programmercarl.com/0242.%E6%9C%89%E6%95%88%E7%9A%84%E5%AD%97%E6%AF%8D%E5%BC%82%E4%BD%8D%E8%AF%8D.html  


```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        record = [0] * 26
        for i in s:
            record[ord(i)-ord('a')]+=1
        for i in t:
            record[ord(i)-ord('a')]-=1
        for i in range(26):
            if record[i]!=0:
                return False
        return True
```

 349. 两个数组的交集 

建议：开始考虑什么时候用set 什么时候用数组，本题其实是使用set的好题，尝试去使用set。

思路：主要是用set去重，然后用get进行字典计数，同时需要记住使用del删掉字典中的元素。

题目链接/文章讲解/视频讲解：https://programmercarl.com/0349.%E4%B8%A4%E4%B8%AA%E6%95%B0%E7%BB%84%E7%9A%84%E4%BA%A4%E9%9B%86.html  


```python
class Solution(object):
    def intersection(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        table={}
        result=set()
        for i in nums1:
            table[i]=table.get(i,0)+1
        for j in nums2:
            if j in table:
                result.add(j)
                del table[j]
        return list(result)
```

 202. 快乐数 

建议：这道题目也是set的应用，其实和上一题差不多，就是套在快乐数一个壳子

思路：这题重点是当new_sum重复出现在set里，就说明要无限循环了，返回False。

题目链接/文章讲解：https://programmercarl.com/0202.%E5%BF%AB%E4%B9%90%E6%95%B0.html  


```python
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        record=set()
        while n not in record:
            record.add(n)
            new_sum=0
            str_n=str(n)
            for i in str_n:
                new_sum+=int(i)**2
            if new_sum==1:
                return True
            else:
                n=new_sum
        return False
```

1. 两数之和 

建议：本题虽然是力扣第一题，但是还是挺难的，也是使用map解决哈希问题的第一题。 

先看视频讲解，然后尝试自己写代码，在看文章讲解，加深印象。 

注意：第一需要把循环过的元素储存在字典里，第二在循环到下一个元素时，判断target-value是否在字典里出现过。字典的key为值，value为index。enumerate拆分index和value。

题目链接/文章讲解/视频讲解：https://programmercarl.com/0001.%E4%B8%A4%E6%95%B0%E4%B9%8B%E5%92%8C.html

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        record=dict()
        for index,value in enumerate(nums):
            key=target-value
            if key in record:
                return [record[key],index]
            record[value]=index
        return []
```
