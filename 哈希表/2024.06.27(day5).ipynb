// https://programmercarl.com/0242.%E6%9C%89%E6%95%88%E7%9A%84%E5%AD%97%E6%AF%8D%E5%BC%82%E4%BD%8D%E8%AF%8D.html  


```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        record = [0] * 26
        for i in s:
            record[ord(i)-ord('a')]+=1
        for i in t:
            record[ord(i)-ord('a')]-=1
        for i in range(26):
            if record[i]!=0:
                return False
        return True
```

```python
class Solution(object):
    def intersection(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: List[int]
        """
        table={}
        result=set()
        for i in nums1:
            table[i]=table.get(i,0)+1
        for j in nums2:
            if j in table:
                result.add(j)
                del table[j]
        return list(result)
```

```python
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """
        record=set()
        while n not in record:
            record.add(n)
            new_sum=0
            str_n=str(n)
            for i in str_n:
                new_sum+=int(i)**2
            if new_sum==1:
                return True
            else:
                n=new_sum
        return False
```

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        record=dict()
        for index,value in enumerate(nums):
            key=target-value
            if key in record:
                return [record[key],index]
            record[value]=index
        return []
```
