找树左下角的值  

本题递归偏难，反而迭代简单属于模板题， 两种方法掌握一下 

题目链接/文章讲解/视频讲解：https://programmercarl.com/0513.%E6%89%BE%E6%A0%91%E5%B7%A6%E4%B8%8B%E8%A7%92%E7%9A%84%E5%80%BC.html

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def findBottomLeftValue(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        res=[]
        if not root:
            return 0
        quene=collections.deque([root])
        while quene:
            size=len(quene)
            temp=[]
            for i in range (size):
                r=quene.popleft()
                temp.append(r.val)
                if r.left:
                    quene.append(r.left)
                if r.right:
                    quene.append(r.right)
            res.append(temp)
        return res[-1][0]
```

路径总和  

本题又一次涉及到回溯的过程，而且回溯的过程隐藏的还挺深,也就是count需要需要回溯到最开始剪掉targetsum-root.val的时候。

112. 路径总和，和113. 路径总和ii一起做了.

题目链接/文章讲解/视频讲解：https://programmercarl.com/0112.%E8%B7%AF%E5%BE%84%E6%80%BB%E5%92%8C.html

```python
# 路径总和112.
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def sum_path(self,root,count):
        # stop signal
        if not root.right and not root.left and count==0:
            return True 
        if not root.right and not root.left:
            return False
        if root.left:
            if self.sum_path(root.left, count-root.left.val):
                return True
        if root.right:
            if self.sum_path(root.right,count-root.right.val):
                return True
    
    def hasPathSum(self, root, targetSum):
        """
        :type root: TreeNode
        :type targetSum: int
        :rtype: bool
        """
        if not root:
            return False
        return self.sum_path(root,targetSum-root.val)

# 路径总和113.
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def sum_path(self,node,count,path,result):
        if not node.left and not node.right and count==0:
            result.append(path[:])
            return

        # left
        if node.left:
            path.append(node.left.val)
            count-=node.left.val 
            self.sum_path(node.left,count,path,result)
            count+=node.left.val #回溯
            path.pop() #回溯
        
        # right
        if node.right:
            path.append(node.right.val)
            count-=node.right.val
            self.sum_path(node.right,count,path,result)
            count+=node.right.val #回溯
            path.pop() #回溯
        return result

        
    def pathSum(self, root, targetSum):
        """
        :type root: TreeNode
        :type targetSum: int
        :rtype: List[List[int]]
        """
        result=[]
        if not root:
            return []
        path=[]
        path.append(root.val)
        return self.sum_path(root,targetSum-root.val,path,result)
        
```

从中序与后序遍历序列构造二叉树 

本题算是比较难的二叉树题目。 

106.从中序与后序遍历序列构造二叉树，105.从前序与中序遍历序列构造二叉树一起做，思路一样的

记住创建支树是TreeNode(val),要点是切割

题目链接/文章讲解/视频讲解：https://programmercarl.com/0106.%E4%BB%8E%E4%B8%AD%E5%BA%8F%E4%B8%8E%E5%90%8E%E5%BA%8F%E9%81%8D%E5%8E%86%E5%BA%8F%E5%88%97%E6%9E%84%E9%80%A0%E4%BA%8C%E5%8F%89%E6%A0%91.html 


```python
# 106.从中序与后序遍历序列构造二叉树
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def buildTree(self, inorder, postorder):
        """
        :type inorder: List[int]
        :type postorder: List[int]
        :rtype: TreeNode
        """
        if not inorder:
            return None
        
        # get root
        root = postorder[-1]
        result=TreeNode(root)

        # change inorder
        root_idex=inorder.index(root)
        inorder_left=inorder[:root_idex]
        inorder_right=inorder[root_idex+1:]

        # change postorder
        postorder_left=postorder[:len(inorder_left)]
        postorder_right=postorder[len(inorder_left):len(postorder)-1]

        # recursion
        # left 
        result.left=self.buildTree(inorder_left,postorder_left)
        # right
        result.right=self.buildTree(inorder_right,postorder_right)
        return result

# 105.从前序与中序遍历序列构造二叉树    
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: TreeNode
        """
        if not inorder:
            return None
        root=preorder[0]
        result=TreeNode(root)

        # change inorder
        root_index=inorder.index(root)
        inorder_left=inorder[:root_index]
        inorder_right=inorder[root_index+1:]

        # change preorder
        preoder_left=preorder[1:len(inorder_left)+1]
        preoder_right=preorder[1+len(inorder_left):]

        # recursion
        result.left=self.buildTree(preoder_left,inorder_left)
        result.right=self.buildTree(preoder_right,inorder_right)
        return result
```
